```python
import yaml
import matplotlib.pyplot as plt
import numpy as np
import os
import tables_io
import pandas as pd
from itables import show
```

```python
RESULTS_DIR = '../results'
SUBMISSIONS = evaluation.get_submissions()
```

```python
data_list = []
for sub_ in SUBMISSIONS:
    cardinal_1_yr_point = os.path.join(RESULTS_DIR, sub_, "taskset_1_cardinal_1yr_point.yaml")
    cardinal_10_yr_point = os.path.join(RESULTS_DIR, sub_, "taskset_1_cardinal_10yr_point.yaml")
    flagship_1_yr_point = os.path.join(RESULTS_DIR, sub_, "taskset_1_flagship_1yr_point.yaml")
    flagship_10_yr_point = os.path.join(RESULTS_DIR, sub_, "taskset_1_flagship_10yr_point.yaml")
    cardinal_1_yr_pit = os.path.join(RESULTS_DIR, sub_, "taskset_1_cardinal_1yr_pit_qq.yaml")
    cardinal_10_yr_pit = os.path.join(RESULTS_DIR, sub_, "taskset_1_cardinal_10yr_pit_qq.yaml")
    flagship_1_yr_pit = os.path.join(RESULTS_DIR, sub_, "taskset_1_flagship_1yr_pit_qq.yaml")
    flagship_10_yr_pit = os.path.join(RESULTS_DIR, sub_, "taskset_1_flagship_10yr_pit_qq.yaml")

    data_cardinal_1_yr_point = yaml.safe_load(open(cardinal_1_yr_point))
    data_cardinal_10_yr_point = yaml.safe_load(open(cardinal_10_yr_point))
    data_flagship_1_yr_point = yaml.safe_load(open(flagship_1_yr_point))
    data_flagship_10_yr_point = yaml.safe_load(open(flagship_10_yr_point))
    data_cardinal_1_yr_pit = yaml.safe_load(open(cardinal_1_yr_pit))
    data_cardinal_10_yr_pit = yaml.safe_load(open(cardinal_10_yr_pit))
    data_flagship_1_yr_pit = yaml.safe_load(open(flagship_1_yr_pit))
    data_flagship_10_yr_pit = yaml.safe_load(open(flagship_10_yr_pit))

    data_dict = dict(submission=sub_)
    data_dict.update(**{f"{k}_1yr": 0.5*(v + data_cardinal_1_yr_point[k]) for k, v in data_flagship_1_yr_point.items()})
    data_dict.update(**{f"{k}_1yr": 0.5*(v + data_cardinal_1_yr_pit[k]) for k, v in data_flagship_1_yr_pit.items()})
    data_dict.update(**{f"{k}_10yr": 0.5*(v + data_cardinal_10_yr_point[k]) for k, v in data_flagship_10_yr_point.items()})
    data_dict.update(**{f"{k}_10yr": 0.5*(v + data_cardinal_10_yr_pit[k]) for k, v in data_flagship_10_yr_pit.items()})

    data_list.append(data_dict)


# Create pivot table
df = pd.DataFrame(data_list)
```
