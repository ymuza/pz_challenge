# Validation plots for pz data challenge (part 2)

#### Standard imports

```python
import os
import matplotlib.pyplot as plt
import numpy as np
import tables_io
from pz_data_challenge import submit_utils
```

#### Data setup and reading files

```python
DATADIR = '../public'
```

```python
TASKSETS = ["taskset_1", "taskset_2"]
SIMS = ["cardinal", "flagship"]
SCENARIOS = ["1yr", "10yr"]
```

```python
data_dict = {}
for taskset_ in TASKSETS:
    for sim_ in SIMS:
        for scenario_ in SCENARIOS:
            key = f"{taskset_}_{sim_}_{scenario_}"
            test_file = os.path.abspath(os.path.join(DATADIR, f"pz_challenge_{taskset_}_{sim_}_test_{scenario_}.hdf5"))
            train_file = os.path.abspath(os.path.join(DATADIR, f"pz_challenge_{taskset_}_{sim_}_training_{scenario_}.hdf5"))
            data_dict[f"{key}_train"] = tables_io.read(train_file)
            data_dict[f"{key}_test"] = tables_io.read(test_file)
```

#### Plotting redshift as a funciton of color v. color

```python
def plot_color_color_redshift(d_dict, l_, b1='g', b2='r', b3='i', sample=1):
    fig = plt.figure()
    data = d_dict[l_]
    color_1 = data[f"mag_{b1}_lsst"] - data[f"mag_{b2}_lsst"]
    color_2 = data[f"mag_{b2}_lsst"] - data[f"mag_{b3}_lsst"]
    _ = plt.scatter(color_1[::sample], color_2[::sample], s=1, c=data['redshift'][::sample], vmin=0, vmax=1.5, cmap='rainbow')
    _ = plt.xlabel(f"{b1} - {b2} [mag]")
    _ = plt.ylabel(f"{b2} - {b3} [mag]")
    _ = plt.xlim(-0.5, 3.0)
    _ = plt.ylim(-0.5, 3.0)
    _ = plt.colorbar(label='z')
    return fig
```

```python
ff = plot_color_color_redshift(data_dict, 'taskset_1_cardinal_10yr_train', sample=1)
ff.savefig('color_color_redshift_taskset_1_cardinal_10yr.png')
```

```python
ff = plot_color_color_redshift(data_dict, 'taskset_1_flagship_10yr_train', sample=10)
ff.savefig('color_color_redshift_taskset_1_flagship_10yr.png')
```

#### Plotting color v. redshift

```python
def plot_color_v_redshift(d_dict, l_, b1='g', b2='r', b3='i', sample=1):
    fig = plt.figure()
    data = d_dict[l_]
    color_1 = data[f"mag_{b1}_lsst"] - data[f"mag_{b2}_lsst"]
    color_2 = data[f"mag_{b2}_lsst"] - data[f"mag_{b3}_lsst"]
    bins = (np.linspace(0., 1.5, 151), np.linspace(-0.5, 3, 71))
    _ = plt.scatter(data['redshift'][::sample], color_1[::sample], label=f"{b1} - {b2}", s=1)
    _ = plt.scatter(data['redshift'][::sample], color_2[::sample], label=f"{b2} - {b3}", s=1)
    _ = plt.xlabel("z")
    _ = plt.ylabel("color [mag]")
    _ = plt.xlim(0, 1.5)
    _ = plt.legend()
    return fig
```
