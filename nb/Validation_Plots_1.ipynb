# Validation plots for pz data challenge (part 1)

#### standard imports

```python
import os
import matplotlib.pyplot as plt
import numpy as np
import tables_io
from pz_data_challenge import submit_utils
```

#### Data setup and reading files

```python
DATADIR = '../public'
```

```python
TASKSETS = ["taskset_1", "taskset_2"]
SIMS = ["cardinal", "flagship"]
SCENARIOS = ["1yr", "10yr"]
```

```python
data_dict = {}
for taskset_ in TASKSETS:
    for sim_ in SIMS:
        for scenario_ in SCENARIOS:
            key = f"{taskset_}_{sim_}_{scenario_}"
            test_file = os.path.abspath(os.path.join(DATADIR, f"pz_challenge_{taskset_}_{sim_}_test_{scenario_}.hdf5"))
            train_file = os.path.abspath(os.path.join(DATADIR, f"pz_challenge_{taskset_}_{sim_}_training_{scenario_}.hdf5"))
            data_dict[f"{key}_train"] = tables_io.read(train_file)
            data_dict[f"{key}_test"] = tables_io.read(test_file)
```

#### plotting the survey footprints

```python
def compare_footprint(d_dict, l1, l2):
    fig = plt.figure()
    _ = plt.scatter(d_dict[l1]['ra'], d_dict[l1]['dec'], label=l1, s=1)   
    _ = plt.scatter(d_dict[l2]['ra'], d_dict[l2]['dec'], label=l2, s=1)
    _ = plt.xlabel('RA [deg]')
    _ = plt.ylabel('DEC [deg]')
    _ = plt.legend()
    return fig
```

```python
fig_footprint_sims = compare_footprint(data_dict, 'taskset_1_cardinal_1yr_train', 'taskset_1_flagship_1yr_train')
```

```python
fig_footprint_test_train = compare_footprint(data_dict, 'taskset_1_cardinal_1yr_train', 'taskset_1_flagship_1yr_test')
```

#### Plotting a histogram of the magnitidues

```python
def plot_i_mags(d_dict, l_list):
    fig = plt.figure()
    bins = np.linspace(16, 26, 101)
    for l_ in l_list:
        _ = plt.hist(d_dict[l_]['mag_i_lsst'], bins=bins, alpha=0.5, label=l_)
    _ = plt.xlabel('i [mag]')
    _ = plt.ylabel('objects / 0.1 mag')
    _ = plt.yscale('log')
    _ = plt.legend()
    return fig
```

```python
labels = ['taskset_1_cardinal_1yr_train', 'taskset_1_cardinal_1yr_test']
fig_mags_taskset1_carindal_1yr = plot_i_mags(data_dict, labels)
```

```python
labels = ['taskset_2_flagship_1yr_train', 'taskset_2_flagship_1yr_test']
fig_mags_taskset2_flagship_1yr = plot_i_mags(data_dict, labels)
```

```python
labels = ['taskset_2_cardinal_1yr_train', 'taskset_2_cardinal_1yr_test']
fig_mags_taskset2_cardinal_1yr = plot_i_mags(data_dict, labels)
```

```python
labels = ['taskset_2_cardinal_10yr_train', 'taskset_2_cardinal_10yr_test']
fig_mags_taskset2_cardinal_10yr = plot_i_mags(data_dict, labels)
```

#### Plotting histograms of the redshifts

```python
def plot_redshifts(d_dict, l_list):
    fig = plt.figure()
    bins = np.linspace(0., 3., 301)
    for l_ in l_list:
        _ = plt.hist(d_dict[l_]['redshift'], bins=bins, alpha=0.5, label=l_)
    _ = plt.xlabel('z')
    _ = plt.ylabel('objects / 0.01')
    _ = plt.yscale('log')
    _ = plt.legend()
    return fig
```

```python
fig_redshifts_taskset1_1yr = plot_redshifts(data_dict, ['taskset_1_cardinal_1yr_train', 'taskset_1_flagship_1yr_train'])
```

```python
fig_redshifts_tasksets_cardinal = plot_redshifts(data_dict, ['taskset_1_cardinal_1yr_train', 'taskset_2_cardinal_1yr_train'])
```

```python
fig_redshifts_tasksets_cardinal = plot_redshifts(data_dict, ['taskset_1_flagship_1yr_train', 'taskset_2_flagship_1yr_train'])
```

#### Plotting the magnitude error v. the magnitude to show the photometric smearing

```python
def plot_mag_v_mag_err(d_dict, l_list):
    fig = plt.figure()
    for l_ in l_list:
        _ = plt.scatter(d_dict[l_]['mag_i_lsst'], d_dict[l_]['mag_i_lsst_err'], label=l_)
    _ = plt.xlabel('i [mag]')
    _ = plt.ylabel(r'$\sigma_i$ [mag]')
    _ = plt.yscale('log')
    _ = plt.legend()
    return fig
```
